# StackAPI

#### Import the necessary libraries here:

```python
import requests

import pandas as pd

import json

from stackapi import StackAPI

import datetime as dt

import re
```

```python
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=1547247600, todate=1577833200, min=20, sort='votes')

print(json.dumps(questions, indent=4))
```

#### Question 1: Find the questions and answers of last month.

##### Creating a DataFrame out of the JSON & isolating the item columns into its own DF

```python
question_dec = pd.DataFrame(questions)

question_dec
```

```python
question_dec['items']
```

```python
questions_items_df = pd.DataFrame(dict(question_dec["items"])).transpose()

questions_items_df.head(20)
```
